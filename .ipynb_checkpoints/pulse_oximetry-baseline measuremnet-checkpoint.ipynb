```python
def znormalize(values):
    value_series = pd.Series(values)
    mean = value_series.mean()
    std = value_series.std()
    return (value_series - mean) / std
```

```python
#Does it make sense to also reindex the column ID's here? Column labels are offset by one. Not sure if this is important or not?
#Paul: It isn't important right now, but it may be important for some analyses, so I went ahead and made that change
all_vals = raw_data[raw_data.columns[4:]]
all_vals.columns = pd.Index([int(column_name) - 1 for column_name in all_vals.columns]) #correct column index offset
all_vals1d = pd.Series(all_vals.values.flatten())
min_val = 70 #-1
all_vals1d_missing_removed = all_vals1d[all_vals1d > min_val]
global_mean = all_vals1d_missing_removed.mean()
global_std = all_vals1d_missing_removed.std()
```

```python
def get_quantile_breakpoints(paa_series, alphabet_size):
    i_vals = np.arange(0, alphabet_size)
    breakpoints = [
        paa_series.quantile(i / float(alphabet_size))
        for i in i_vals
    ] + [np.inf]
    return breakpoints
```

```python
def piecewise_aggregate_approximation(series, paa_size):
    paa_size = int(paa_size)
    series_length = int(series.count())
    #take care of the special case where there is no dimensionality reduction
    if series_length == paa_size:
        paa = series
    else:
        if series_length % paa_size == 0:
            paa = series.reshape([series_length / paa_size, paa_size], order="F").mean(axis=0)
        else:
            temp = pd.DataFrame(series).T.reindex(index = xrange(paa_size), method="ffill")
            expanded_sub_section = temp.values.reshape(paa_size * series_length, order="F")
            paa = expanded_sub_section.reshape([series_length, paa_size], order="F").mean(axis=0)
    return pd.Series(paa)
```

```python
#For the cuts, we may wish to bound asymmetrically, such that we have a floor of ?70% below which our signal is considered 'bad'; the functional range is realy above 80%
#Paul: should I try cutting those off? It may be good to slice those signals out when we filter out the -1s
def get_cuts(paa_series, breakpoints):
    cuts = paa_series.apply(lambda paa_element: np.sum(breakpoints <= paa_element))
    return cuts
```

```python
def convert_cuts_to_sax(cuts):
    sax_series = cuts.apply(lambda cut: string.ascii_lowercase[cut - 1])
    sax_str = "".join(sax_series.values)
    return (sax_series, sax_str)
```

```python
def replace_missing_patient_vals(patientid):
    vals = pd.Series(raw_data.ix[patientid].values[4:])
    missing_values = vals[vals < min_val]
    missing_values_index = missing_values.index
    missing_vals_removed = vals.drop(missing_values_index)
    missing_vals_removed_index = missing_vals_removed.index
    
    min_index = missing_vals_removed_index[0]
    max_index = missing_vals_removed_index[-1]
    interpolation_func = interp1d(x = missing_vals_removed_index,
                                  y = missing_vals_removed)
    #Interpolation is just to fill in missing indices and SpO2 measurements I think?
    #Paul: Yes, that is correct. There are two methods below that I tried. The interpolated_vals method does a 
    #linear interpolation to overwrite spans of repeated -1 values, and the vals_reindexed method stretches
    #the last value before the -1 span and repeats it until the next value (eg 87 84 -1 -1 -1 90 -> 87 84 84 84 84 90)
    #Both methods seem to show the same behavior 
    interpolated_vals = pd.Series(interpolation_func(np.arange(min_index, max_index + 1)))
    #vals_reindexed = missing_vals_removed.reindex(np.arange(min_index, max_index), method="ffill")
    #See above about  normalization here; might need to skew this to set a floor of 70% or so;

    return interpolated_vals
```

## sax calculation

```python
all_points = []

for patientid in raw_data.index:
    interpolated_vals = replace_missing_patient_vals(patientid)
    all_points += list(interpolated_vals.values)
    
all_points_series = pd.Series.from_array(all_points)
```

```python
#PT: If I understand this correctly, we are compressing 10-second epochs into a single frame? 
#Paul: Yes, that is correct. paa_ticks_to_compress below controls that quantity

patient_saxes = pd.Series(index = raw_data.index)
patient_paas = {}
#paa_ticks_to_compress determines how many points to include per frame when reducing 
#dimensionality. this is conceptually similar to moving average, but with the averaging window
#stepping forward by paa_ticks_to_compress points instead of 1 point. its slightly more 
#complicated than that, but thats the basic idea (see piecewise_aggregate_approximation function)
paa_ticks_to_compress = 10
sax_alphabet_size = 10
for patientid in raw_data.index:
    interpolated_vals = replace_missing_patient_vals(patientid)
    paa = piecewise_aggregate_approximation(interpolated_vals, interpolated_vals.count() / paa_ticks_to_compress) 
    patient_paas[patientid] = paa

all_paas = pd.Series(pd.DataFrame.from_dict(patient_paas).values.flatten())
breakpoints = get_quantile_breakpoints(all_paas, sax_alphabet_size)

for patientid in raw_data.index:    
    paa = patient_paas[patientid]
    #breakpoints = get_normal_breakpoints(paa, sax_alphabet_size)
    cuts = get_cuts(paa, breakpoints)
    (sax_series, sax_str) = convert_cuts_to_sax(cuts)
    patient_saxes.ix[patientid] = sax_str

patient_saxes
#very nice code, although that's coming from an amateur
#Paul: Thank you!
```

```python
def get_raw_sequence_motif_vector(sequence, alphabet_size, motif_length = 2):
    l = motif_length
    a = alphabet_size
    n = a ** l #possible motif combinations
    Mp = np.zeros(n) #motif vector
    for motif_pos in range(len(sequence) - l + 1):
        motif = sequence[motif_pos:motif_pos + l] #current motif
        i = 0
        for j in np.arange(l):
            #where is get_letter_index from? Silly question, but how does this work if it comes at the bottom of this cell?
            #Paul: these functions can refer to each other since they are being called from later in the notebook and are both stored in Python's memory
            kj = get_letter_index(motif[j])
            i += kj * a ** (l - j - 1)
        all_motifs[i] = motif
        Mp[i] += 1
    return pd.Series(Mp)

def get_letter_index(letter):
    return string.lowercase.index(letter)
```

```python
#This reports, for each subject (in rows), how often each 3-letter motif occurs, given a 5-letter alphabet? 125 columns = 5**3 possible combinations for motifs?
#Paul: Yes, that is correct
all_motifs = {}
motif_length = 2
patient_motif_counts = patient_saxes.apply(get_raw_sequence_motif_vector, 
                                           alphabet_size = sax_alphabet_size,
                                           motif_length = motif_length)
```

```python
icon = (patient_motif_counts / patient_motif_counts.max()).var().fillna(0)
```
