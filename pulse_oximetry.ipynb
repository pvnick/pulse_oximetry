```python
top_k = {} #variance: location
num_items = 0
lowest_held_variance = 0
k = 50

def insert(top_k, key, value):
    if key not in top_k:
        top_k[key] = []
    top_k[key].append(value)
    return top_k
    
flat_data = pd.Series.from_csv("/home/pvnick/oximetry.txt", index_col=None)
n = flat_data.count();
m = 500
num_windows = n - m
for i in range(num_windows):
    subsequence = flat_data[i:i+m]
    variance = subsequence.var()
    if num_items < k:
        num_items += 1
        top_k = insert(top_k, variance, i)
        if variance > lowest_held_variance:
            lowest_held_variance = variance
    elif variance > lowest_held_variance:
        del top_k[lowest_held_variance]
        top_k = insert(top_k, variance, i)
        lowest_held_variance = np.min(top_k.keys())
```

```python
def get_quantile_breakpoints(paa_series, alphabet_size):
    i_vals = np.arange(0, alphabet_size)
    breakpoints = [
        paa_series.quantile(i / float(alphabet_size))
        for i in i_vals
    ] + [np.inf]
    return breakpoints

def get_normal_breakpoints(paa_series, alphabet_size):
    mean = paa_series.mean()
    stddev = paa_series.std()
    alphabet_size = int(alphabet_size)
    i_vals = np.arange(0, alphabet_size)
    breakpoints = [sc.stats.norm.ppf(q=i / float(alphabet_size), loc = mean, scale = stddev) 
                   for i in range(alphabet_size + 1)]
    return breakpoints
```

```python
def piecewise_aggregate_approximation(series, paa_size):
    paa_size = int(paa_size)
    series_length = int(series.count())
    #take care of the special case where there is no dimensionality reduction
    if series_length == paa_size:
        paa = series
    else:
        if series_length % paa_size == 0:
            paa = series.reshape([series_length / paa_size, paa_size], order="F").mean(axis=0)
        else:
            temp = pd.DataFrame(series).T.reindex(index = xrange(paa_size), method="ffill")
            expanded_sub_section = temp.values.reshape(paa_size * series_length, order="F")
            paa = expanded_sub_section.reshape([series_length, paa_size], order="F").mean(axis=0)
    return pd.Series(paa)
```

```python
#For the cuts, we may wish to bound asymmetrically, such that we have a floor of ?70% below which our signal is considered 'bad'; the functional range is realy above 80%
#Paul: should I try cutting those off? It may be good to slice those signals out when we filter out the -1s
def get_cuts(paa_series, breakpoints):
    cuts = paa_series.apply(lambda paa_element: np.sum(breakpoints <= paa_element))
    return cuts
```

```python
def convert_cuts_to_sax(cuts):
    sax_series = cuts.apply(lambda cut: string.ascii_lowercase[cut - 1])
    sax_str = "".join(sax_series.values)
    return (sax_series, sax_str)
```

```python
#This selects a random number of patients, number defined by sample size, with replacement (bootstrapping)
#sums the motif counts of those sampled patients on a per-motif basis, then sums this vector which serves as a denominator
#for normalization; ulitmatley returns a vector of normalized motif importances via bootstrapping
def sample_motif_importances(sample_size):
    sample_indices = np.random.choice(patient_motif_counts.index, size=sample_size, replace=True)
    sample = patient_motif_counts.ix[sample_indices]
    individual_motif_counts = sample.sum()
    total_motif_quantities = individual_motif_counts.sum(axis=1)
    x = individual_motif_counts / total_motif_quantities
    return x
```

```python
def get_shape(df):
    delta_sign = np.sign(df)
    #replace zeros with last sign
    shape = delta_sign.drop(delta_sign[delta_sign==0].index).reindex(delta_sign.index, method="ffill")
    return shape

shape = delta.apply(get_shape, axis=1)
```

```python
def get_features(patient_id):
    sax_str = patient_saxes.ix[patient_id]
    sax_letter_indices = pd.Series(list(sax_str)).apply(lambda sax_letter: string.lowercase.index(sax_letter))
    delta = sax_letter_indices.T.diff().T
    deltadelta = delta.T.diff().T
    shape = get_shape(delta)
    return pd.DataFrame.from_dict({
        "sax": sax_letter_indices,
        "delta": delta,
        "deltadelta": deltadelta,
        "shape": shape
    })
```
