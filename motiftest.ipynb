```python
def znorm(series):
    mean = series.mean()
    std = series.std()
    return (series - mean) / std

def plot_motif(row):
    plt.figure(figsize=[10, 5])
    plt.hold(True)
    plt.plot(np.arange(int(row[0])),
             znorm(data_series[int(row[1]):int(row[1]) + int(row[0])]),
             np.arange(int(row[0])),
             znorm(data_series[int(row[2]):int(row[2]) + int(row[0])])
             )
    plt.hold(False)
    plt.show()
motifs.apply(plot_motif, axis=1)
```
